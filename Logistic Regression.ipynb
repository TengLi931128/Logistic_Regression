```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def sigmoid(x):
    return 1/(1+np.exp(-x))

def forward(W,B,X):
    Z = np.dot(W.T,X)+B
    Y_hat = sigmoid(Z)
    return Y_hat

def backward(Y_hat,Y,W,B,X):
    #gradient
    dZ = Y_hat-Y
    dW = np.dot(X,dZ.T)/L
    dB = np.sum(dZ)/L
    #update weights
    W = W-alpha*dW
    B = B-alpha*dB
    return W,B

def predict(X,W,B):
    Z = np.dot(W.T,X)+B
    C = np.sign(sigmoid(Z)-0.5)
    return C
    
```

```python
#Dataset: 5000 entities with 2 features
L = 5000
X = np.random.uniform(-100,100,[2,L])
Y = np.sign(X[0]-X[1])
alpha = 0.9
epoch = 200

#Weights initialize
dW = np.zeros([2,1])
dB = 0
W = np.random.rand(2,1)
B = np.random.normal()
```

```python
#Training
for i in range(epoch):
    #forward
    Y_hat = forward(W,B,X)
    #backward
    W,B = backward(Y_hat,Y,W,B,X)
print(W)
print(B)
```

```python
# 3D graph
x = X[0]
y = X[1]
z = Y
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x,y,z)
ax.set_zlabel('Z',fontdict={'size':15,'color':'blue'})
ax.set_ylabel('Y',fontdict={'size':15,'color':'yellow'})
ax.set_xlabel('X',fontdict={'size':15,'color':'red'})
plt.show()
```
